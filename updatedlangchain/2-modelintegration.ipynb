# Streaming and Batch

### Streaming
Most models can stream their output content while it is being generated . By displaying output progressively , streaming significantly improves user experience, particularly for longer responses. calling stream() returns an iterator that yeilds output chunks as they are produced . you can use a loop to process each chunk in real-time:

```python
#normally it shows result after its done processing 
model.invoke("who are you?")
```

```python
# using streaming
for chunk in model.stream("write me an essay on Artificial Intelligence"):
    print(chunk.text, end=" ",flush=True)
```

### Batch 
Batching is a collection of independent requests to a model can significantly improve performance and reduce costs , as the processing can be done in parallel:

```python
responses = model.batch([
    "why do parrots have colorful feathers?",
    "how do airplane fly?",
    "why am i single ?"
])
for response in responses:
    print(response)
```

```python
#max_concurrency

model.batch([
    "why do parrots have colorful feathers?",
    "how do airplane fly?",
    "why am i single ?"],
config={
    'max_concurrency':5, # limit to 5 parallel calls
}
)
```
